```python
# Carga inicial de librer√≠as
import pandas as pd
import numpy as np
from datetime import datetime
from statistics import mean
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.decomposition import PCA
from plotnine import ggplot, aes, geom_line, labs
```

```python
# Cargamos los datos de las estaciones de cada una de las islas
datos = pd.read_csv('../data/Final/datos_final.csv', header=0)

datos_Tenerife
datos_GranCanaria
datos_LaPalma
datos_ElHierro
datos_LaGomera
datos_Lanzarote
datos_Fuerteventura
```

```python
islas = datos['isla'].unique()
num_islas = len(islas)
num_cols = 2  
num_rows = round (num_islas / num_cols)

fig, axes = plt.subplots(num_rows, num_cols, figsize=(12, 5*num_rows))

for i, isla in enumerate(islas):
    datos_isla = datos[datos['isla'] == isla]
    ax = axes[i // num_cols, i % num_cols] if num_rows > 1 else axes[i % num_cols]
    ax.plot(datos_isla['fecha'], datos_isla['numero_casos'], label = isla)
    ax.set_title(f'Isla de {isla}')
    ax.set_xlabel('Fecha')
    ax.set_ylabel('N√∫mero de casos')
    ax.tick_params(axis='x', rotation=90)
    ax.set_xticks=datos_isla['fecha'].iloc[::100]


plt.tight_layout()

plt.show()
```

```python
islas = datos['isla'].unique()

fig, axis = plt.subplots(4, 2, figsize=(12, 18))  # Ajust√© el tama√±o para mejor visualizaci√≥n

for i, isla in enumerate(islas):
    datos_isla = datos[datos['isla'] == isla]    
    
    # Usando plot en vez de scatter para dibujar l√≠neas
    axis[i//2, i%2].plot(datos_isla['fecha'], datos_isla['numero_casos'])
    axis[i//2, i%2].set_title('N√∫mero de casos isla de: {}'.format(isla))
    
    axis[i//2, i%2].figure.autofmt_xdate()

# Ajustamos el dise√±o y mostramos el gr√°fico
plt.tight_layout()
plt.show()
```

```python
valores_islas = datos['isla'].unique()

# Configurar el tama√±o de la figura y el n√∫mero de subgr√°ficos
num_subplots = len(valores_islas)
fig, axes = plt.subplots(num_subplots, 1, figsize=(10, 6*num_subplots), sharex=True)

# Iterar sobre las islas y trazar la serie temporal correspondiente para cada una en un subgr√°fico diferente
for i, isla in enumerate(valores_islas):
    datos_isla = datos[datos['isla'] == isla] 
    axes[i].plot(datos_isla['fecha'], datos_isla['numero_casos'], label = isla)  # Trazar la serie temporal para la isla actual en el subgr√°fico correspondiente
    axes[i].set_title(isla)  # Establecer el t√≠tulo del subgr√°fico como el nombre de la isla
    axes[i].set_ylabel('Valor')  # Establecer la etiqueta del eje y

    
# A√±adir etiqueta del eje x al √∫ltimo subgr√°fico
axes[-1].set_xlabel('Fecha')
axes[-1].set_ylabel('N√∫mero de casos')

# Ajustar el dise√±o de los subgr√°ficos para evitar superposiciones
plt.tight_layout()

# Mostrar la gr√°fica
plt.show()
```

```python
response = 'numero_casos'

columnas = datos.columns

num_feats = [columna for columna in columnas if columna != 'isla' and columna != 'fecha']
```

```python
datos_corr = datos.select_dtypes(include=['number'])
correlacion = datos_corr.corrwith(datos_corr[response])
```

```python
plt.figure(figsize=(10, 6))
correlacion.plot(kind='bar', color='skyblue')
plt.title('Correlaci√≥n de la variable objetivo con las variables num√©ricas')
plt.xlabel('Variables Num√©ricas')
plt.ylabel('Coeficiente de Correlaci√≥n')
plt.xticks(rotation=90)
plt.show()
```

```python
for i in num_feats:
    print("Correlaci√≥n entre los atributos ", response, "-", i, ":", datos[response].corr(datos[i], method='pearson'))
```
