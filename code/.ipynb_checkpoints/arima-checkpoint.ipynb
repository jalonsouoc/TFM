```python
# Carga inicial de librerías
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import STL
from statsmodels.graphics.tsaplots import plot_acf
```

```python
def estudio_STL(df):
    stl = STL(df, period=12)
    res = stl.fit()
    
    fig = res.plot()
    
    plt.show()
```

```python
# Cargamos los datos de las estaciones de cada una de las islas
datos = pd.read_csv('../data/Final/datos_final.csv', header=0)
```

```python
datos_stl = datos [datos['isla'] == 'TENERIFE']
datos_stl = datos_stl[['fecha', 'numero_casos']]
datos_stl.set_index('fecha', inplace=True)
```

```python
print(datos_stl)
```

```python
estudio_STL(datos_stl)
```

```python
datos_1d = datos_stl
datos_1d = datos_1d['numero_casos'].diff()
```

```python
estudio_STL(datos_1d)
```

```python
datos_2d = datos_stl
datos_2d = datos_2d['numero_casos'].diff().diff()
```

```python
estudio_STL(datos_2d)
```

```python
datos_3d = datos_stl
datos_3d = datos_3d['numero_casos'].diff().diff().diff()
```

```python
estudio_STL(datos_3d)
```

# ACF

```python
plot_acf(datos['numero_casos'], lags=160)
plt.xlabel('Lag')
plt.ylabel('Autocorrelación')
plt.title('Autocorrelación de la Serie Temporal')
plt.show()
```
